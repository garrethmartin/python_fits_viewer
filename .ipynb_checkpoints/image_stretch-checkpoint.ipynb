```python
%matplotlib notebook
from fits_mask_tools import FitsViewer, MaskPainter
```

### The FitsViewer loads a .fits image and allows rescaling via the dropdown menu and sliders. We can also save a snapshot toa .png file.

```python
crop = (2250,3250,1250,2250)
fv = FitsViewer("calexp-HSC-I-9813-5,3.fits", hdu_index=1, crop=crop)
```

### The MaskPainter takes the current state of the FitsViewer and allows us to paint multiple masks over the image.

```python
painter = MaskPainter(fv, brush_size=10)
```

### Finally, we can pass the mask file back to FitsViewer to view only the masked pixels.

```python
fv_masked = FitsViewer("calexp-HSC-I-9813-5,3.fits", hdu_index=1, crop=crop, mask_file='./masks/mask_000.fits')
```

### We can also retrieve the maked data directly and do whatever we want with it.

```python
import matplotlib.pyplot as plt
import numpy as np

masked_array = fv_masked.masked_data_cropped

fig, ax = plt.subplots(figsize=(6,6))
ax.set_axis_off()

im = ax.imshow(np.log10(masked_array), origin='lower', cmap='jet', interpolation='nearest')
plt.show()
```
